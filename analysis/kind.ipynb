```python
import re
import unicodedata
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize

# ====== 텍스트 전처리 ======
def normalize_title(s: str) -> str:
    s = unicodedata.normalize("NFKC", str(s)).lower()
    s = re.sub(r"[·ㆍ•‧∙]", " ", s)       # 특수기호 통일
    s = re.sub(r"[\(\[\{][^\)\]\}]*[\)\]\}]", " ", s)  # 괄호내용 제거
    s = re.sub(r"[^0-9a-z가-힣\s]", " ", s) # 한글/영문/숫자/공백만
    s = re.sub(r"\s+", " ", s).strip()
    return s

# ====== 메인 함수 ======
def cluster_titles(df: pd.DataFrame, threshold: float = 0.33, min_similarity: float = 0.4,
                   ngram_min: int = 2, ngram_max: int = 5):
    """
    df: 반드시 df["title"], df["count"] 컬럼 필요 (count 없으면 value_counts로 만들어서 넣어주세요)
    threshold: 군집화 거리 임계값 (낮을수록 더 세밀하게 분리)
    min_similarity: 라벨 매칭 최소 유사도 (0~1)
    """
    titles = df["title"].astype(str).tolist()
    counts = df["count"].fillna(1).astype(int).to_numpy()

    # ---- 1) 벡터화 ----
    norm_titles = [normalize_title(t) for t in titles]
    vec = TfidfVectorizer(analyzer="char_wb", ngram_range=(ngram_min, ngram_max))
    X = vec.fit_transform(norm_titles)

    # ---- 2) 거리행렬 & 군집 ----
    S = cosine_similarity(X)
    D = 1.0 - S
    clustering = AgglomerativeClustering(
        n_clusters=None, linkage="average", metric="precomputed",
        distance_threshold=threshold
    )
    labels = clustering.fit_predict(D)

    # ---- 3) 대표 라벨 선택 ----
    rep_idx_by_cluster = {}
    rows = []
    for cid in sorted(set(labels)):
        idxs = np.where(labels == cid)[0]
        Xc = X[idxs]
        wc = counts[idxs].astype(float)
        wc = wc / (wc.sum() if wc.sum() > 0 else 1)
        
        # 가중 평균 centroid 계산을 더 안전하게 처리
        # Xc를 dense array로 변환 후 계산
        Xc_dense = Xc.toarray() if hasattr(Xc, 'toarray') else Xc
        centroid = np.average(Xc_dense, axis=0, weights=wc)
        
        # centroid가 1차원이면 2차원으로 변환
        if centroid.ndim == 1:
            centroid = centroid.reshape(1, -1)
        centroid = normalize(centroid, norm="l2")
        sims = cosine_similarity(Xc, centroid).ravel()
        best_idx = idxs[np.argmax(sims)]
        rep_idx_by_cluster[cid] = best_idx
        rows.append({
            "cluster_id": cid,
            "base_label": titles[best_idx],
            "size": len(idxs),
            "examples": " | ".join([titles[i] for i in idxs[:3]])
        })
    base_df = pd.DataFrame(rows).sort_values("size", ascending=False).reset_index(drop=True)

    # ---- 4) 매핑 ----
    base_labels = [titles[rep_idx_by_cluster[cid]] for cid in sorted(rep_idx_by_cluster)]
    base_X = X[list(rep_idx_by_cluster.values())]
    S_map = cosine_similarity(X, base_X)
    best_idx = S_map.argmax(axis=1)
    best_sim = S_map.max(axis=1)

    matched_labels, matched_cids = [], []
    for i, sim in enumerate(best_sim):
        j = int(best_idx[i])
        if sim >= min_similarity:
            matched_labels.append(base_labels[j])
            matched_cids.append(j)
        else:
            matched_labels.append("UNKNOWN")
            matched_cids.append(-1)

    map_df = df.copy()
    map_df["cluster_id"] = matched_cids
    map_df["base_label"] = matched_labels
    map_df["cosine_sim"] = best_sim

    return base_df, map_df
```

```python
# 수정된 함수로 클러스터링 테스트
df_test = df["title"].value_counts().reset_index()
df_test.columns = ["title","count"]

print(f"테스트할 제목 수: {len(df_test)}")
print("클러스터링 시작...")
base_df, map_df = cluster_titles(df_test)
print("클러스터링 완료!")
print(f"총 클러스터 수: {len(base_df)}")
base_df
```

```python
from dotenv import load_dotenv
import os
import requests

load_dotenv()
DART_API_KEY = os.getenv("DART_API_KEY")

def get_convertible_bond(corp_code, bgn_de, end_de):
    """
    전환사채권 발행결정 데이터를 조회하는 함수
    
    Parameters:
    corp_code (str): 공시대상회사의 고유번호 (8자리)
    bgn_de (str): 시작일 (YYYYMMDD 형식)
    end_de (str): 종료일 (YYYYMMDD 형식)
    
    Returns:
    dict: API 응답 결과
    """
    
    url = "https://opendart.fss.or.kr/api/cvbdIsDecsn.json"
    
    # 요청 파라미터
    params = {
        'crtfc_key': DART_API_KEY,
        'corp_code': corp_code,
        'bgn_de': bgn_de,
        'end_de': end_de
    }
    
    try:
        # API 요청 수행
        response = requests.get(url, params=params)
        response.raise_for_status()
        
        # JSON 응답 반환
        data = response.json()
        
        # 상태 확인
        if data.get('status') == '000':
            print(f"✅ 데이터 조회 성공: {len(data.get('list', []))}건")
            return data
        else:
            print(f"❌ 에러 발생: {data.get('message')}")
            return data
            
    except requests.exceptions.RequestException as e:
        print(f"❌ 네트워크 오류: {e}")
        return None
    except Exception as e:
        print(f"❌ 예상치 못한 오류: {e}")
        return None
```

```python
def format_convertible_bond_info(data):
    """전환사채 발행정보를 포맷팅하여 출력하는 함수"""
    if not data or data.get('status') != '000' or not data.get('list'):
        print("❌ 데이터가 없습니다.")
        return
    
    bond_info = data['list'][0]
    
    # 숫자 포맷팅 함수 (억원 단위로 변환)
    def format_amount(amount_str):
        if not amount_str or amount_str == '-':
            return '-'
        try:
            amount = int(amount_str)
            return f"{amount:,}원 ({amount/100000000:.1f}억)"
        except:
            return amount_str
    
    # 퍼센트 포맷팅 함수
    def format_percent(percent_str):
        if not percent_str or percent_str == '-':
            return '-'
        return f"{percent_str}%"
    
    print(f"📊 전환사채 발행정보")
    print(f"=" * 50)
    print(f"회사명: {bond_info.get('corp_name', '-')}")
    print(f"접수번호: {bond_info.get('rcept_no', '-')}")
    print(f"이사회결의일: {bond_info.get('bddd', '-')}")
    print()
    
    print(f"💰 발행정보")
    print(f"사채종류: {bond_info.get('bd_knd', '-')}")
    print(f"발행금액: {format_amount(bond_info.get('bd_fta'))}")
    print(f"발행방법: {bond_info.get('bdis_mthn', '-')}")
    print(f"사채만기일: {bond_info.get('bd_mtd', '-')}")
    print()
    
    print(f"📈 이자율 정보")
    print(f"표면이자율: {format_percent(bond_info.get('bd_intr_ex'))}")
    print(f"만기이자율: {format_percent(bond_info.get('bd_intr_sf'))}")
    print()
    
    print(f"🔄 전환 정보")
    print(f"전환가액: {format_amount(bond_info.get('cv_prc'))} (1주당)")
    print(f"전환비율: {format_percent(bond_info.get('cv_rt'))}")
    # print(f"전환주식수: {bond_info.get('cvisstk_cnt', '-'):,}주" if bond_info.get('cvisstk_cnt', '-') != '-' else "전환주식수: -")
    print(f"주식총수 대비: {format_percent(bond_info.get('cvisstk_tisstk_vs'))}")
    print(f"전환청구기간: {bond_info.get('cvrqpd_bgd', '-')} ~ {bond_info.get('cvrqpd_edd', '-')}")
    print(f"최저조정가액: {format_amount(bond_info.get('act_mktprcfl_cvprc_lwtrsprc'))}")
    print()
    
    print(f"📅 일정")
    print(f"청약일: {bond_info.get('sbd', '-')}")
    print(f"납입일: {bond_info.get('pymd', '-')}")
    print()
    
    # 자금조달 목적이 있는 경우만 출력
    funding_purposes = []
    if bond_info.get('fdpp_fclt') not in [None, '-', '']:
        funding_purposes.append(f"시설자금: {format_amount(bond_info.get('fdpp_fclt'))}")
    if bond_info.get('fdpp_op') not in [None, '-', '']:
        funding_purposes.append(f"운영자금: {format_amount(bond_info.get('fdpp_op'))}")
    if bond_info.get('fdpp_dtrp') not in [None, '-', '']:
        funding_purposes.append(f"채무상환자금: {format_amount(bond_info.get('fdpp_dtrp'))}")
    if bond_info.get('fdpp_ocsa') not in [None, '-', '']:
        funding_purposes.append(f"타법인증권취득자금: {format_amount(bond_info.get('fdpp_ocsa'))}")
    if bond_info.get('fdpp_etc') not in [None, '-', '']:
        funding_purposes.append(f"기타자금: {format_amount(bond_info.get('fdpp_etc'))}")
    
    if funding_purposes:
        print(f"💼 자금조달 목적")
        for purpose in funding_purposes:
            print(f"{purpose}")
        print()
    
    print(f"ℹ️ 기타정보")
    if bond_info.get('rpmcmp') not in [None, '-', '']:
        print(f"대표주관회사: {bond_info.get('rpmcmp', '-')}")
    if bond_info.get('grint') not in [None, '-', '']:
        print(f"보증기관: {bond_info.get('grint', '-')}")
    print(f"증권신고서 제출대상: {bond_info.get('rs_sm_atn', '-')}")
    if bond_info.get('ex_sm_r') not in [None, '-', '']:
        print(f"제출면제사유: {bond_info.get('ex_sm_r', '-')}")


```
