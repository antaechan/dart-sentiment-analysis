```python
# null인 ret 값들을 가진 종목들의 OHLCV 데이터 확인
import os
from pathlib import Path

# null ret 값을 가진 행들 필터링
null_ret_df = df[
    (df[['ret_1m', 'ret_3m', 'ret_10m', 'ret_15m', 'ret_30m', 'ret_60m']].isna().all(axis=1)) &
    (pd.to_datetime(df['disclosed_at']).dt.tz_convert('Asia/Seoul').dt.dayofweek.isin([0,1,2,3,4])) &
    (pd.to_datetime(df['disclosed_at']).dt.tz_convert('Asia/Seoul').dt.strftime('%Y%m%d').isin(target_dates))
]

print(f"null ret 값을 가진 공시 건수: {len(null_ret_df)}")

# 각 종목별로 OHLCV 파일 존재 여부 확인
kospi_ohlcv_dir = Path("/Volumes/APFS/KOSPI/ohlcv")
kosdaq_ohlcv_dir = Path("/Volumes/APFS/KOSDAQ/ohlcv")

missing_files = []
existing_files = []

for idx, row in null_ret_df.iterrows():
    stock_code = row['stock_code']
    market = row['market']
    disclosed_date = pd.to_datetime(row['disclosed_at']).tz_convert('Asia/Seoul').strftime('%Y%m%d')
    
    # 시장에 따라 디렉터리 선택
    if market == 'KOSPI':
        ohlcv_dir = kospi_ohlcv_dir
    elif market == 'KOSDAQ':
        ohlcv_dir = kosdaq_ohlcv_dir
    else:
        continue
    
    # 해당 날짜의 OHLCV 파일 경로
    ohlcv_file = ohlcv_dir / f"{disclosed_date}_1m.parquet"
    
    if ohlcv_file.exists():
        # 파일이 존재하면 해당 종목의 데이터가 있는지 확인
        try:
            ohlcv_data = pl.read_parquet(ohlcv_file)
            # 종목코드를 'code' 컬럼으로 변경 (파일 구조에 맞게)
            stock_data = ohlcv_data.filter(pl.col("종목코드") == stock_code)
            
            if len(stock_data) > 0:
                existing_files.append({
                    'id': row['id'],
                    'stock_code': stock_code,
                    'market': market,
                    'disclosed_date': disclosed_date,
                    'company_name': row['company_name'],
                    'ohlcv_records': len(stock_data)
                })
                print(f"✓ {market} {stock_code} ({row['company_name']}) - {disclosed_date}: {len(stock_data)}개 레코드")
            else:
                missing_files.append({
                    'id': row['id'],
                    'stock_code': stock_code,
                    'market': market,
                    'disclosed_date': disclosed_date,
                    'company_name': row['company_name'],
                    'reason': '파일은 존재하지만 해당 종목 데이터 없음'
                })
                print(f"✗ {market} {stock_code} ({row['company_name']}) - {disclosed_date}: 파일은 있지만 종목 데이터 없음")
        except Exception as e:
            missing_files.append({
                'id': row['id'],
                'stock_code': stock_code,
                'market': market,
                'disclosed_date': disclosed_date,
                'company_name': row['company_name'],
                'reason': f'파일 읽기 오류: {str(e)}'
            })
            print(f"✗ {market} {stock_code} ({row['company_name']}) - {disclosed_date}: 파일 읽기 오류")
    else:
        missing_files.append({
            'id': row['id'],
            'stock_code': stock_code,
            'market': market,
            'disclosed_date': disclosed_date,
            'company_name': row['company_name'],
            'reason': 'OHLCV 파일 없음'
        })
        print(f"✗ {market} {stock_code} ({row['company_name']}) - {disclosed_date}: OHLCV 파일 없음")

print(f"\n=== 요약 ===")
print(f"OHLCV 데이터가 존재하는 종목: {len(existing_files)}개")
print(f"OHLCV 데이터가 없는 종목: {len(missing_files)}개")

# 데이터가 존재하는 종목들 상세 정보
if existing_files:
    print(f"\n=== OHLCV 데이터가 존재하는 종목들 ===")
    existing_df = pd.DataFrame(existing_files)
    print(existing_df)

# 데이터가 없는 종목들 상세 정보
if missing_files:
    print(f"\n=== OHLCV 데이터가 없는 종목들 ===")
    missing_df = pd.DataFrame(missing_files)
    print(missing_df.groupby('reason').size())
```

```python
# OHLCV 대신 partitioned 폴더에서 데이터 존재 여부 확인
print("=== Partitioned 폴더에서 데이터 존재 여부 확인 ===")

KOSPI_PARTITION_DIR = Path("/Volumes/APFS/KOSPI/partitioned")
KOSDAQ_PARTITION_DIR = Path("/Volumes/APFS/KOSDAQ/partitioned")

import polars as pl
from pathlib import Path

existing_partition_files = []
missing_partition_files = []

# null ret 값을 가진 행들 필터링
null_returns_df = df[
    (df[['ret_1m', 'ret_3m', 'ret_10m', 'ret_15m', 'ret_30m', 'ret_60m']].isna().all(axis=1)) &
    (pd.to_datetime(df['disclosed_at']).dt.tz_convert('Asia/Seoul').dt.dayofweek.isin([0,1,2,3,4])) &
    (pd.to_datetime(df['disclosed_at']).dt.tz_convert('Asia/Seoul').dt.strftime('%Y%m%d').isin(target_dates))
]

for _, row in null_returns_df.iterrows():
    stock_code = row['stock_code']
    market = row['market']
    # disclosed_at이 이미 문자열인 경우 처리
    if isinstance(row['disclosed_at'], str):
        disclosed_date = pd.to_datetime(row['disclosed_at']).strftime('%Y%m%d')
    else:
        disclosed_date = row['disclosed_at'].strftime('%Y%m%d')
    
    # 시장별 파티션 디렉터리 설정
    if market == 'KOSPI':
        partition_dir = KOSPI_PARTITION_DIR
    elif market == 'KOSDAQ':
        partition_dir = KOSDAQ_PARTITION_DIR
    else:
        missing_partition_files.append({
            'id': row['id'],
            'stock_code': stock_code,
            'market': market,
            'disclosed_date': disclosed_date,
            'company_name': row['company_name'],
            'reason': '알 수 없는 시장'
        })
        continue
    
    # 파티션 파일 경로 확인
    partition_path = partition_dir / f"체결일자={disclosed_date}"
    
    if partition_path.exists():
        try:
            # 해당 종목의 데이터가 있는지 확인
            df_partition = pl.scan_parquet(
                partition_path,
                hive_partitioning=True,
                low_memory=True,
            ).filter(
                (pl.col("체결일자") == int(disclosed_date)) &
                (pl.col("종목코드") == stock_code)
            ).select(["종목코드", "체결일자"]).limit(1).collect()
            
            if len(df_partition) > 0:
                existing_partition_files.append({
                    'id': row['id'],
                    'stock_code': stock_code,
                    'market': market,
                    'disclosed_date': disclosed_date,
                    'company_name': row['company_name']
                })
                print(f"✓ {market} {stock_code} ({row['company_name']}) - {disclosed_date}: 파티션 데이터 존재")
            else:
                missing_partition_files.append({
                    'id': row['id'],
                    'stock_code': stock_code,
                    'market': market,
                    'disclosed_date': disclosed_date,
                    'company_name': row['company_name'],
                    'reason': '파티션 폴더는 존재하지만 해당 종목 데이터 없음'
                })
                print(f"✗ {market} {stock_code} ({row['company_name']}) - {disclosed_date}: 파티션 폴더는 있지만 종목 데이터 없음")
        except Exception as e:
            missing_partition_files.append({
                'id': row['id'],
                'stock_code': stock_code,
                'market': market,
                'disclosed_date': disclosed_date,
                'company_name': row['company_name'],
                'reason': f'파티션 파일 읽기 오류: {str(e)}'
            })
            print(f"✗ {market} {stock_code} ({row['company_name']}) - {disclosed_date}: 파티션 파일 읽기 오류")
    else:
        missing_partition_files.append({
            'id': row['id'],
            'stock_code': stock_code,
            'market': market,
            'disclosed_date': disclosed_date,
            'company_name': row['company_name'],
            'reason': '파티션 폴더 없음'
        })
        print(f"✗ {market} {stock_code} ({row['company_name']}) - {disclosed_date}: 파티션 폴더 없음")

print(f"\n=== 파티션 데이터 요약 ===")
print(f"파티션 데이터가 존재하는 종목: {len(existing_partition_files)}개")
print(f"파티션 데이터가 없는 종목: {len(missing_partition_files)}개")

# 파티션 데이터가 존재하는 종목들 상세 정보
if existing_partition_files:
    print(f"\n=== 파티션 데이터가 존재하는 종목들 ===")
    existing_partition_df = pd.DataFrame(existing_partition_files)
    print(existing_partition_df)

# 파티션 데이터가 없는 종목들 상세 정보
if missing_partition_files:
    print(f"\n=== 파티션 데이터가 없는 종목들 ===")
    missing_partition_df = pd.DataFrame(missing_partition_files)
    print(missing_partition_df.groupby('reason').size())
```
